```python
import requests
import json

# DeepSeek API endpoint and your API key
api_url = "https://api.deepseek.com/v1/chat/completions"
api_key = "sk-bedcb944ad3d4bae9b9dfa4080af75ce"

# Headers for authorization and content
headers = {
    "Authorization": f"Bearer {api_key}",
    "Content-Type": "application/json"
}

# The payload (system + user messages)
payload = {
    "model": "deepseek-chat",
    "temperature": 0.6,
    "messages": [
        {
            "role": "system",
            "content": (
                "You are a professional travel planner writing detailed and impressive itineraries for distinguished academics. "
                "Your tone is polished, respectful, and intellectually elegant. Be precise and structured."
            )
        },
        {
            "role": "user",
            "content": (
                "Write a 3-day itinerary for Professor Arun, a distinguished academic visiting Bogota for an international research conference. "
                "Include time slots, meaningful cultural and historical experiences, academic engagements, and high-end dining options. "
                "Ensure the itinerary is well-structured and suitable for a person of high intellectual stature."
            )
        }
    ]
}

# Make the API request
response = requests.post(api_url, headers=headers, json=payload)

# Parse and display the result
if response.status_code == 200:
    result = response.json()
    itinerary_text = result["choices"][0]["message"]["content"]
    print("=== Itinerary for Professor Arun ===\n")
    print(itinerary_text)
else:
    print("Error:", response.status_code)
    print(response.text)
```

```python
import streamlit as st
import requests

# ========== CONFIGURATION ==========
API_KEY = "sk-bedcb944ad3d4bae9b9dfa4080af75ce"
API_URL = "https://api.deepseek.com/v1/chat/completions"

# ========== SESSION STATE ==========
if "messages" not in st.session_state:
    st.session_state.messages = [
        {
            "role": "system",
            "content": (
                "You are a respectful, insightful travel assistant helping Professor Arun navigate his 3-day academic trip to Kyoto. "
                "You provide elegant answers about itinerary items, meetings, transport, cultural visits, and fine dining."
            )
        }
    ]

# ========== APP HEADER ==========
st.title("Professor Arun's Bogota Itinerary Assistant")
st.markdown("Ask any questions about your visit â€” academic sessions, cultural highlights, fine dining, and more.")

# ========== USER INPUT ==========
user_input = st.text_input("Your question:", placeholder="e.g., What's on the agenda for Day 2 evening?")

if user_input:
    st.session_state.messages.append({"role": "user", "content": user_input})

    response = requests.post(
        API_URL,
        headers={
            "Authorization": f"Bearer {API_KEY}",
            "Content-Type": "application/json"
        },
        json={
            "model": "deepseek-chat",
            "temperature": 0.5,
            "messages": st.session_state.messages
        }
    )

    if response.status_code == 200:
        assistant_reply = response.json()["choices"][0]["message"]["content"]
        st.session_state.messages.append({"role": "assistant", "content": assistant_reply})
    else:
        st.error(f"API Error: {response.status_code}")
        assistant_reply = ""

# ========== CHAT HISTORY ==========
for msg in st.session_state.messages[1:]:  # Skip system message
    if msg["role"] == "user":
        st.markdown(f"**Professor Arun:** {msg['content']}")
    else:
        st.markdown(f"**Assistant:** {msg['content']}")
```
